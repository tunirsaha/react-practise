# Variable Basics


```python
# Variable Types

a="Hello"
b='k'
c=10
d=10.5
e=-123
f=True
g=None

print(type(a), type(b),type(c), type(d),type(e), type(f),type(g))

# Pythonic Version
values = ["Hello", 'k', 10, 10.5, -123, True, None]
print(*(type(v) for v in values))

# Use of * - unpacking operator
h=[1,2,3]
print(h, *h)
```

```python
# Variable Assignment

a,b,c=1,2,3
print('1:',a,b,c)
# Multi assign

a=c=10
print('2:',a,b,c)
# Equi assign

c=20
print('3:',a,b,c)
# No object references like JS
```

```python
# Typecasting

print(float('1.14'), int(1.14), int(float('1.14')))
print('Bool(-0.0001): ', bool(-0.0001),'Bool(' '): ', bool(' '), 'Bool(''): ', bool(''), "Bool('0'): ", bool('0'), "Bool('1'): ", bool('1'))

# Pythonic Version: F Strings
tests = [-0.0001, " ", "", "0", "1"]
for t in tests:
    print(f"{t!r} -> {bool(t)}")

# value!r = repr(value)
print('Empty Space', '')
print('Empty Space', f"{''!r}")

print(type(True), type('True'))
```

# Mathematical

```python
# Object Reference not applicable

a=10
b=5
a+=b
b+=a
print(a,b)
```

```python
# Bitwise

a=8
b=3
print(bin(5), bin(3), bin(a & b), a & b) # AND

# Pythonic Version: Format Specifiers
print(f"{5:b}", f"{3:b}", f"{a & b:b}", a & b)

a=5
print(a^5, a^5^5) # XOR

a=6
print(~a) # NOT
```

# Logical Operator

```python
# Truth and Truthy, False and Falsy

print(bool(''),''==False) # Empty strings are falsy but not false
print(bool('abc'), 'abc'==True) # strings are truthy but not true

print(bool(0), bool(0.0)) # zero is falsy and false
print(0==False, 0.0==False)

print(bool(1), bool(1.0), bool(-123)) # one is truthy and true
print(1==True, 1.0==True, 1.5==True)

print(bool(None)) # None is falsy but not true/false
print(None==True, None==False)
```

```python
# Logical on Group of Operators

print('abc' and 'def' and 'efg') # if all truthy, return last value
print(0 and False and 'tunir' and 0) # if first falsy, return first value

print('abc' or 'def' or 'xyz' or False) # if all truthy, return first value
print(0 or False or 2 or 'tunir' or 1) # if all false, return first truthy
```

# Problems

```python
'''
You are given a total number of seconds.
Convert it to hours, minutes, and seconds and print it in the format hh:mm:ss.
Sample Input: 3661
Sample Output: 01:01:01
'''

total_seconds = int(input())
hours = total_seconds//3600 #extract hours
minutes = (total_seconds%3600)//60 #remove hours, extract minutes
seconds = total_seconds%60 #remove all minutes
output = f"{hours:02d}:{minutes:02d}:{seconds:02d}"
print(output)
```

```python
# Swapping Numbers

a = 10
b = 20
print('A :', a, 'B :',b)
print('A :', id(a), 'B :',id(b))
a, b = b, a
print('A :', a, 'B :',b)
```

```python
a = int(input('Enter Number: '))
match a:
  case a if a%2==0: print('Even')
  case a if a%2!=0: print('Odd')
  case _: print('Error') # Wildcard for default
```

```python
x,y=int(input('Enter X: ')),int(input('Enter Y: '))
match x,y:
  case x,y if x>y:
    print('X is greater')
  case x,y if y>x:
    print('Y is greater')
```

```python
a = int(input('Enter Num: '))
match a:
  case 1 | 2 | 3: print('Low')
  case 4 | 5 | 6: print('Medium')
  case 7 | 8 | 9: print('High')
  case a if a>9: print('Very High')
```

```python
a=''
for ch in 'abcdaeafg':
  a += ch if ch != 'a' else '*' # ternary if attached to first statement
print(a)

# Cleaner Pythonic version
a = ''.join('*' if ch == 'a' else ch for ch in 'abcdaeafg')
print(a)
```
